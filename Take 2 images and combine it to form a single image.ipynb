```python
#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Get 1st img
img = cv2.imread('linux.jpg')
```

```python
#Display 1st img
plt.imshow(img)
```

```python
#Display shape of img
img.shape
```

```python
#Get 2nd img
img2 = cv2.imread('ubuntu.png')
```

```python
#Display 2nd img
plt.imshow(img2)
```

```python
#Display shape of img
img2.shape
```

```python
#For getting same size of both imgs we have to crop the 2nd img
crop = img2[300:550 , 210:980]
```

```python
#Display 2nd img after cropped
plt.title('crop1')
plt.imshow(crop)
```

```python
#Display shape of cropped img
crop.shape
```

```python
#Merge both imgs
merge = np.hstack(( img , crop))
```

```python
#Display merged img
plt.imshow(merge)
```
